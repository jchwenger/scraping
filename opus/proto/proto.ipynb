```python
import opustools
```

```python
import inspect
```

```python
og = opustools.opus_get.OpusGet(source='fr', preprocess='mono', list_resources=True)
```

```python
og.get_files()
```

```python
corpora, file_n, total_size = og.get_corpora_data()
```

```python
corpora
```

```python
no_tok = []
file_n = 0
total_size = 0
for c in corpora:
    if 'tok' not in c['url']:
        no_tok.append(c)
        file_n += 1
        total_size += c['size']
```

```python
no_tok
```

```python
print(file_n, 'files')
print(f'total size: {total_size:,}')
```

```python
og.download(no_tok, file_n, str(total_size))
```
